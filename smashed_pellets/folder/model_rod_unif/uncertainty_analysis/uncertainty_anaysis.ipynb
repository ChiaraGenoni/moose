```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

path = 'CT4.csv'
list = []

df = pd.read_csv(path)
time = df['time']

# Define a prefix for column selection
prefix_600 = 'CT4_600'
prefix_800 = 'CT4_800'
prefix_1000 = 'CT4_1000'

# Create a list of columns with headers starting with the specified prefix
selected_columns_600 = [col for col in df.columns if col.startswith(prefix_600)]
selected_columns_800 = [col for col in df.columns if col.startswith(prefix_800)]
selected_columns_1000 = [col for col in df.columns if col.startswith(prefix_1000)]


df_600 = df[selected_columns_600]/1e6
df_800 = df[selected_columns_800]/1e6
df_1000 = df[selected_columns_1000]/1e6


df_600['mean_value'] = df_600.mean(axis=1)
df_600['min_value'] = df_600.min(axis=1)
df_600['max_value'] = df_600.max(axis=1)
df_600 = pd.concat([time, df_600], axis=1)
plt.plot(df_600['time'],df_600['mean_value'])

df_800['mean_value'] = df_800.mean(axis=1)
df_800['min_value'] = df_800.min(axis=1)
df_800['max_value'] = df_800.max(axis=1)
df_800 = pd.concat([time, df_800], axis=1)
plt.plot(df_800['time'],df_800['mean_value'])

df_1000['mean_value'] = df_1000.mean(axis=1)
df_1000['min_value'] = df_1000.min(axis=1)
df_1000['max_value'] = df_1000.max(axis=1)
df_1000 = pd.concat([time, df_1000], axis=1)
plt.plot(df_1000['time'],df_1000['mean_value'])

# Plot the error band
plt.fill_between(time, df_600['min_value'], df_600['max_value'], color='blue', alpha=0.2, label='4.3 MPa')
plt.fill_between(time, df_800['min_value'], df_800['max_value'], color='red', alpha=0.2, label='5.6 MPa')
plt.fill_between(time, df_1000['min_value'], df_1000['max_value'], color='green', alpha=0.2, label='7.0 MPa')
# Add labels and title
plt.xlabel('Time (s)')
plt.ylabel('Pressure (MPa)')
plt.grid()
plt.title('CT4')
plt.legend()

```
