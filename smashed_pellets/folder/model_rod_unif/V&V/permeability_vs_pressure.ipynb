```python
import matplotlib.pyplot as plt

# Sample data (replace this with your own data)
pressure_data = [600, 800, 1000]  # Pressure values (e.g., in PSI)
K1 = [7.4e-13,  6.6e-12,  3.3e-13] # Permeability values (e.g., in Darcies)
K2 = [3.1e-10, 2.5e-10, 2.5e-13]
K = [5.3e-14, 2.0e-14, 1.3e-14]

# Plot the data
plt.plot(pressure_data, K1, marker='o', linestyle='-', label = "K1")
plt.plot(pressure_data, K, marker='o', linestyle='-', label = "K")
plt.xlabel('Pressure (psi)')
plt.ylabel('Permeability')
plt.title('Permeability as a function of Pressure')
plt.legend() 
plt.grid(True)
plt.show()

# Plot the data
plt.plot(pressure_data, K2, marker='o', linestyle='-', label = "K2")
plt.xlabel('Pressure (psi)')
plt.ylabel('Permeability')
plt.title('Permeability as a function of Pressure')
plt.legend()
plt.grid(True)
plt.show()
```
