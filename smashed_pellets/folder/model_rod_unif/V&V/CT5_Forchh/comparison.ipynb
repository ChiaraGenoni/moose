```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

best_fit_800 = pd.read_csv('model_pellets_800.csv')
best_fit_1000 = pd.read_csv('model_pellets_1000.csv')
data20 = pd.read_csv('experimental_data.csv')

plt.plot(data20['Time3'], data20['P_3']/1e6, 'v', color='tab:red' , label = 'Experimental 7.0 MPa')
plt.plot(best_fit_1000['time'], best_fit_1000['inlet-p']/1e6, linestyle = '--', color='tab:orange', label = 'CFD 7.0 MPa')

plt.plot(data20['Time4'], data20['P_4']/1e6, 'v', color='tab:green' , label = 'Experimental 5.6 MPa')
plt.plot(best_fit_800['time'], best_fit_800['inlet-p']/1e6, linestyle = '--', color='tab:olive', label = 'CFD 5.6 MPa')
plt.xlabel('Time (s)', fontsize = 12)
plt.xlim(0,1200)
plt.ylim(0, 8) 
plt.ylabel('Pressure (MPa)', fontsize = 12)
plt.tick_params(axis='both', which='major', labelsize=12) 
plt.legend(fontsize = 12) 
plt.grid() 
```
