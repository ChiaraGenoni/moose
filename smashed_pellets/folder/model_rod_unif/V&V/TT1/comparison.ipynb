```python
# Function to compute Absolute Root Mean Square Error (ARMSE) 
def compute_rmse(y_values, experimental_y_values):
    absolute_error = np.abs(y_values - experimental_y_values)
    squared_error = absolute_error**2
    mean_squared_error = np.mean(squared_error)
    rmse = np.sqrt(mean_squared_error)
    return rmse


y_values = Forchh_600_1['inlet-p']/1e6
x_values = Forchh_600_1['time']
experimental_y_values = np.interp(x_values, experimental['time'], experimental['TT1_600_2']/1e6)
range_of_observed_values = np.max(experimental_y_values) - np.min(experimental_y_values)
rmse = compute_rmse(y_values, experimental_y_values)
print("rmse_600_Forchh ", rmse)

y_values = Forchh_600_2['inlet-p']/1e6
x_values = Forchh_600_2['time']
experimental_y_values = np.interp(x_values, experimental['time'], experimental['TT1_600_1']/1e6)
range_of_observed_values = np.max(experimental_y_values) - np.min(experimental_y_values)
rmse = compute_rmse(y_values, experimental_y_values)
print("rmse_600_Forchh ", rmse)

y_values = Darcy_600_1['inlet-p']/1e6
x_values = Darcy_600_1['time']
experimental_y_values = np.interp(x_values, experimental['time'], experimental['TT1_600_2']/1e6)
range_of_observed_values = np.max(experimental_y_values) - np.min(experimental_y_values)
rmse = compute_rmse(y_values, experimental_y_values)
print("rmse_600_Darcy ", rmse)

y_values = Darcy_600_2['inlet-p']/1e6
x_values = Darcy_600_2['time']
experimental_y_values = np.interp(x_values, experimental['time'], experimental['TT1_600_1']/1e6)
range_of_observed_values = np.max(experimental_y_values) - np.min(experimental_y_values)
rmse = compute_rmse(y_values, experimental_y_values)
print("rmse_600_Darcy ", rmse)







y_values = Forchh_800_1['inlet-p']/1e6
x_values = Forchh_800_1['time']
experimental_y_values = np.interp(x_values, experimental['time'], experimental['TT1_800_1']/1e6)
range_of_observed_values = np.max(experimental_y_values) - np.min(experimental_y_values)
rmse = compute_rmse(y_values, experimental_y_values)
print("rmse_800_Forchh ", rmse)

y_values = Forchh_800_2['inlet-p']/1e6
x_values = Forchh_800_2['time']
experimental_y_values = np.interp(x_values, experimental['time'], experimental['TT1_800_2']/1e6)
range_of_observed_values = np.max(experimental_y_values) - np.min(experimental_y_values)
rmse = compute_rmse(y_values, experimental_y_values)
print("rmse_800_Forchh ", rmse)

y_values = Darcy_800_1['inlet-p']/1e6
x_values = Darcy_800_1['time']
experimental_y_values = np.interp(x_values, experimental['time'], experimental['TT1_800_2']/1e6)
range_of_observed_values = np.max(experimental_y_values) - np.min(experimental_y_values)
rmse = compute_rmse(y_values, experimental_y_values)
print("rmse_800_Darcy ", rmse)

y_values = Darcy_800_2['inlet-p']/1e6
x_values = Darcy_800_2['time']
experimental_y_values = np.interp(x_values, experimental['time'], experimental['TT1_800_1']/1e6)
range_of_observed_values = np.max(experimental_y_values) - np.min(experimental_y_values)
rmse = compute_rmse(y_values, experimental_y_values)
print("rmse_800_Darcy ", rmse)






y_values = Forchh_1000_1['inlet-p']/1e6
x_values = Forchh_1000_1['time']
experimental_y_values = np.interp(x_values, experimental['time'], experimental['TT1_1000_1']/1e6)
range_of_observed_values = np.max(experimental_y_values) - np.min(experimental_y_values)
rmse = compute_rmse(y_values, experimental_y_values)
print("rmse_1000_Forchh ", rmse)

# y_values = Forchh_1000_2['inlet-p']/1e6
# x_values = Forchh_800_2['time']
# experimental_y_values = np.interp(x_values, experimental['time'], experimental['TT1_800_2']/1e6)
# range_of_observed_values = np.max(experimental_y_values) - np.min(experimental_y_values)
# rmse = compute_rmse(y_values, experimental_y_values)
# print("rmse_800_Forchh ", rmse)

y_values = Darcy_1000_1['inlet-p']/1e6
x_values = Darcy_1000_1['time']
experimental_y_values = np.interp(x_values, experimental['time'], experimental['TT1_1000_2']/1e6)
range_of_observed_values = np.max(experimental_y_values) - np.min(experimental_y_values)
rmse = compute_rmse(y_values, experimental_y_values)
print("rmse_1000_Darcy ", rmse)

# y_values = Darcy_800_2['inlet-p']/1e6
# x_values = Darcy_800_2['time']
# experimental_y_values = np.interp(x_values, experimental['time'], experimental['TT1_800_1']/1e6)
# range_of_observed_values = np.max(experimental_y_values) - np.min(experimental_y_values)
# rmse = compute_rmse(y_values, experimental_y_values)
# print("rmse_800_Darcy ", rmse)
```
